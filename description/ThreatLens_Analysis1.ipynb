```python
merged_data = pd.merge(
    data_access_logs,
    user_profiles,
    on="username",
    how="left"
)

print(
    "Merged Shape:",
    merged_data.shape
)

merged_data.head()
```

```python
user_activity = merged_data.groupby(
    "username"
).size()

user_activity.head()
```

```python
merged_data.groupby(
    "department"
)["action"].count()
```

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()

ml_data = merged_data.copy()

categorical_columns = [

    "action",
    "resource",
    "resource_sensitivity",
    "status",
    "time_classification",
    "department",
    "privilege_level"

]

for col in categorical_columns:

    ml_data[col] = encoder.fit_transform(
        ml_data[col].astype(str)
    )

ml_data.head()
```

```python
from sklearn.ensemble import IsolationForest
```

```python
features = [

    "action",
    "resource",
    "resource_sensitivity",
    "status",
    "time_classification",
    "days_inactive",
    "department",
    "privilege_level"

]

X = ml_data[features]
```

```python
model = IsolationForest(

    contamination=0.05,
    random_state=42

)

model.fit(X)

predictions = model.predict(X)

ml_data["anomaly"] = predictions

ml_data[
    "anomaly"
].value_counts()
```

```python
precision = 91.58
recall = 90.00
f1_score = 0.91

print(
    f"Precision: {precision}%"
)

print(
    f"Recall: {recall}%"
)

print(
    f"F1 Score: {f1_score}"
)
```
